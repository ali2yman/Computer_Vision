## how to open video using ur cam

```python
cap = cv2.VideoCapture(0)
while(1):
    ret,frame = cap.read()
    if ret:
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow("original",frame)
        cv2.imshow("gray",gray)
        if cv2.waitKey(10)&0XFF == ord("q"):
            break

cap.release()
cv2.destroyAllWindows()          # then destroy all the opend window
```

## color tracking 

```python
# this is how to know the green range and isoliate the other parts of the img

cap = cv2.VideoCapture(0)
while(1):
    ret,frame = cap.read()
    if ret:
        hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
#         lower and higher limit for green 
        lower_limit = np.array([50,50,50])
        upper_limit = np.array([70,255,255])
# Create a binary mask using cv2.inRange() to identify the pixels within the specified color range (green).
        mask = cv2.inRange(hsv,lower_limit,upper_limit)
        res = cv2.bitwise_and(frame,frame,mask=mask)
        
        cv2.imshow("mask",mask)   #hna by5ly ay green white we elba2y black
        cv2.imshow("original",frame)
        cv2.imshow("result",res)   #hna by5ly ay green green we elba2y black
        
        if cv2.waitKey(10)&0XFF == ord("q"):
            break

cap.release()
cv2.destroyAllWindows()          # then destroy all the opend window
```
