```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

<B>Binary Thresholding<B>

```python
img = cv2.imread('cloud.jpeg',0)
ret,th = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,th2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)

plt.figure(figsize=(16,12))
plt.subplot(131),plt.imshow(img,'gray'),plt.title("original")
plt.subplot(132),plt.imshow(th,"gray"),plt.title("binary")
plt.subplot(133),plt.imshow(th2,"gray"),plt.title("binary inverse")
```

<B>Adptive thresholding (Mean - Gaussian)<B>
    
adaptive thresholding, specifically mean and Gaussian adaptive thresholding, is another image segmentation technique commonly used in computer vision and image processing. Unlike global thresholding methods like Otsu's method, adaptive thresholding computes local thresholds for smaller regions of an image, making it more effective for images with non-uniform illumination or varying contrasts.

In mean adaptive thresholding, a threshold value is calculated for each pixel based on the mean of the pixel values in its neighborhood. Similarly, in Gaussian adaptive thresholding, the threshold value is determined based on a weighted sum of the neighborhood values, where the weights are determined by a Gaussian window.

```python
th = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)   # 11 >> for the mask - 2 >> constant ntr7o mn el result 3shan sa3t byb2a fyh shwyt noise
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)   # 11 >> for the mask - 2 >> constant ntr7o mn el result 3shan sa3t byb2a fyh shwyt noise

plt.figure(figsize=(16,12))
plt.subplot(131),plt.imshow(img,'gray'),plt.title("original")
plt.subplot(132),plt.imshow(th,"gray"),plt.title("ADAPTIVE_THRESH_MEAN")
plt.subplot(133),plt.imshow(th2,"gray"),plt.title("ADAPTIVE_THRESH_GAUSSIAN")
```

<b>OTSU Thresholding<B>
    
The Otsu's thresholding algorithm computes the optimal threshold value by maximizing the between-class variance of the pixel intensities, effectively separating the pixels into two classes: foreground and background. This approach is particularly useful when there is a bimodal distribution of pixel intensities in the image, where the intensities are clustered into two distinct groups representing the foreground and the background.

The Otsu's method is beneficial for tasks such as image segmentation, object recognition, and feature extraction, where the separation of objects from the background is essential. It is commonly used in various image processing applications, such as medical image analysis, document processing, and industrial quality control, among others.

```python
ret,th = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)  

plt.figure(figsize=(16,12))
plt.subplot(131),plt.imshow(img,'gray'),plt.title("original")
plt.subplot(132),plt.imshow(th,"gray"),plt.title("THRESH_OTSU")
```
