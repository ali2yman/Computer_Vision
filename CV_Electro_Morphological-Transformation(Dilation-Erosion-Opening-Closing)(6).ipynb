```python
import cv2 
import numpy as np  
import matplotlib.pyplot as plt 
```

```python
img  = cv2.imread("MorphologicalTransformation.png",0)
ret,th = cv2.threshold(img,190,255,cv2.THRESH_BINARY)

plt.imshow(th,"gray")
```

<B>Erosion<B>
    
 It removes pixels on object boundaries and shrinks the size of objects. It is used to separate connected objects.

```python
kernel = np.ones((5,5))
erosion = cv2.erode(th,kernel,iterations=2)
plt.imshow(erosion,"gray")
```

<B>Dilation<B>

It adds pixels to the boundaries of objects in an image, increasing the size of the objects. It is used to merge nearby objects.

```python
kernel = np.ones((5,5))
dilation = cv2.dilate(th,kernel,iterations=4)
plt.imshow(dilation,"gray")
```

```python
img  = cv2.imread("finger-noise.png",0)
ret,th = cv2.threshold(img,190,255,cv2.THRESH_BINARY)

plt.imshow(th,"gray")
```

<B>Opening<B>

It is an erosion followed by dilation. It is useful in removing noise from an image.

```python
opening = cv2.morphologyEx(th,cv2.MORPH_OPEN,kernel)
plt.imshow(opening,"gray")
```

<b>Closing<B>


It is a dilation followed by erosion. It is useful in closing small holes or gaps in an object.

```python
closing = cv2.morphologyEx(th,cv2.MORPH_CLOSE,kernel)

plt.imshow(closing,"gray")
```

<b>Morphological gradient<B>


is another operation in morphological image processing. It is a useful tool for edge detection, which essentially highlights the boundaries of objects in an image. The morphological gradient is the difference between the dilation and the erosion of an image. Mathematically, it can be represented as:

Morphological Gradient=Dilation(f)Erosion(f)

```python
img  = cv2.imread("cloud.JPEG",0)
ret,th = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

plt.imshow(th,"gray")
```

```python
gradient = cv2.morphologyEx(th,cv2.MORPH_GRADIENT,kernel)
plt.imshow(gradient,"gray")
```
