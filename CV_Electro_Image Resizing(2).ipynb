```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
img  = cv2.imread("1879542-39587833-2560-1440.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

hight = img.shape[0]
weidth = img.shape[1]

resized  = cv2.resize(img,(weidth*5,hight*5),interpolation = cv2.INTER_LINEAR)
```

```python
plt.imshow(img)
plt.show()
print(img.shape)
```

```python
plt.figure(figsize=(16,10))
plt.imshow(resized)
plt.show()
print(resized.shape)
```
